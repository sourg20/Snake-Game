```python
import pygame
import sys
import random

# Initialize pygame
pygame.init()

# Create the clock
clock = pygame.time.Clock()

# Create the screen
screen = pygame.display.set_mode((600, 400))

# Set the background color
screen.fill((255, 255, 255))

# Create the snake
snake = [(100, 50), (90, 50), (80, 50)]

# Create the food
food = (300, 300)

# Set the snake's direction
direction = "right"

# Start the game loop
while True:
    # Limit the frame rate to 15 FPS
    clock.tick(15)

    # Handle events
    for event in pygame.event.get():
        # Quit if the user quits
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

        # Change the snake's direction if the user presses a key
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP and direction != "down":
                direction = "up"
            elif event.key == pygame.K_DOWN and direction != "up":
                direction = "down"
            elif event.key == pygame.K_LEFT and direction != "right":
                direction = "left"
            elif event.key == pygame.K_RIGHT and direction != "left":
                direction = "right"

    # Move the snake
    head = snake[0]
    if direction == "up":
        head = (head[0], head[1] - 10)
    elif direction == "down":
        head = (head[0], head[1] + 10)
    elif direction == "left":
        head = (head[0] - 10, head[1])
    elif direction == "right":
        head = (head[0] + 10, head[1])

    # Check for collisions with boundaries
    if head[0] < 0 or head[0] >= 600 or head[1] < 0 or head[1] >= 400:
        pygame.quit()
        sys.exit()

    # Check for collisions with itself
    if head in snake:
        pygame.quit()
        sys.exit()

    snake.insert(0, head)  # Add the new head to the snake

    # Check if the snake has eaten the food
    if head == food:
        # Generate a new food location on a grid
        while food in snake:
            food = (
                random.randint(0, 59) * 10,
                random.randint(0, 39) * 10,
            )
    else:
        # Remove the tail segment of the snake
        snake.pop()

    # Clear the screen
    screen.fill((255, 255, 255))

    # Draw the snake
    for segment in snake:
        pygame.draw.rect(screen, (0, 0, 0), (*segment, 10, 10))

    # Draw the food
    pygame.draw.rect(screen, (255, 0, 0), (*food, 10, 10))

    # Update the display
    pygame.display.update()
```
